```python
import zedstat 
import pylab as plt
plt.style.use('ggplot')
zt=zedstat.processRoc(df=pd.read_csv('/home/ishanu/ZED/Research/pub_alzheimer_/data_2022/ROC/TRUVEN_PROBABLE_65-999_basic_F.csv',index_col=0),
           order=3, 
           total_samples=502+470,
           positive_samples=470,
           alpha=0.01,
           prevalence=470/(502+470))
zt.smooth(STEP=0.001,interpolate=False)
zt.allmeasures(interpolate=True)
#print('-----')
#display(zt.get())
zt.usample(precision=3)
#print('-----xxxxx')
#display(zt.get())
zt.getBounds()
#print('-----wwwww')
#display(zt.get())

ax=plt.gca()
ax=zt.get().tpr.plot(ax=ax,color='k')
plt.fill_between(x=zt.get().index, y1=zt.df_lim['U'].tpr,y2=zt.df_lim['L'].tpr, color='k',alpha=.2)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_ylabel('tpr');
```

```python
import numpy as np
from scipy.interpolate import UnivariateSpline
def grpfunc(grp):

    n = len(grp)
    x = np.arange(n)

    spl = UnivariateSpline(x, grp.values, s=n)

    return spl(x)
```

```python
xf1=xf.reset_index().groupby('fpr').max()
from scipy.interpolate import UnivariateSpline
xf2=xf1.copy()
xf2.ppv=UnivariateSpline(xf1.index.values,xf1.ppv.values,k=1,s=None)(xf1.index.values)

xf1.set_index('tpr').ppv.plot()
xf2.set_index('tpr').ppv.plot()
```

```python
xf2['LR+']=UnivariateSpline(xf1.index.values,xf1['LR+'].values,k=1,s=None)(xf1.index.values)
xf2.set_index('tpr')['LR+'].plot()
```
